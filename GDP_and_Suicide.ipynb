```python
def run_simulation(system):
    """Runs the constant growth model.
    
    system: System object
    
    returns: TimeSeries
    """
    results = TimeSeries()
    results[system.t_0] = system.t_0
    
    for t in linrange(system.t_0, system.t_end):
        adding = system.a + system.annual_growth
        results[t+1] = adding
    
    return results
```
